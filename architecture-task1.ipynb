```python
dictionary = {'Hello' : 'Hi', 'Bye' : 'Goodbye', 'List' : 'Array'}
value = input()
result = None
for key in dictionary:
    if dictionary[key] == value:
        result = key
print(result)
```

```python
str_list = ['abc', 'bcd', 'abc', 'abd', 'abd', 'dcd', 'abc']
result_dict = {}
for s in str_list:
    if s in result_dict.keys():
        result_dict[s] += 1
    else:
        result_dict[s] = 1
print(list(result_dict.values()))
```
